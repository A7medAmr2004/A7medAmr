```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label_En'] = le.fit_transform(df['label'])
print(df[['label', 'label_En']].value_counts().sort_values(ascending=False))
```

```python
num_col=list(df.select_dtypes(include=['int64','float64']).columns)
num_col
```

```python
q1 = df[num_col].quantile(0.25)
q3 = df[num_col].quantile(0.75)
iqr = q3 - q1
uper_fence = q3 + 1.5 * iqr
lower_fence = q1 - 1.5 * iqr
```

```python
def include_outliers(df=df,num_cols=num_col,lower_fence=lower_fence,uper_fence=uper_fence):
    exist_outlier=dict()
    for col in num_cols:
        if ((df[col]>uper_fence[col]).any()) or  ((df[col]<lower_fence[col]).any()):
            exist_outlier[col]=True

    return exist_outlier
```

```python
outlier_dict=include_outliers()
outlier_dict
```

```python
for col in outlier_dict.keys():
    df[col]=np.where(df[col]>uper_fence[col],uper_fence[col],df[col])
    df[col]=np.where(df[col]<lower_fence[col],lower_fence[col],df[col])
```

```python
plt.figure(figsize=(10,8))
for i,col in enumerate(outlier_dict.keys()):
    plt.subplot(2,3,i+1)
    sns.boxplot(df,y=col)
    plt.title(col)

plt.suptitle('Box plot for each col')
plt.tight_layout()
plt.show()
```

# kMeans

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('number of Clusters')
plt.ylabel('Inertia')
plt.title('determine the best number of Clusters by using Elbow Method')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=42)  
df_scaled['Cluster'] = kmeans.fit_predict(df_scaled)

print(df_scaled.head()) 
```

```python
df_scaled.head()
```

```python
print(df_scaled['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_features = pca.fit_transform(df_scaled.drop(columns=['Cluster']))

df_scaled['PCA1'] = pca_features[:, 0]
df_scaled['PCA2'] = pca_features[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue=df_scaled['Cluster'], palette='viridis', data=df_scaled)
plt.title(f'Clusters Visualization (K = {optimal_k})')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(pca_features,df_scaled['Cluster'])
print(f"Silhouette Score: {score}")
```

# Logistic Regression

```python
x = df.drop(columns=['label_En'])
y = df['label_En']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")


```

```python
from sklearn.metrics import classification_report , confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# DBScan

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt



neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(x)
distances, indices = nbrs.kneighbors(x)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:, 1]  


plt.plot(distances)
plt.xlabel('Sample Index')
plt.ylabel('Distance to Nearest Neighbor')
plt.title('Distances to Nearest Neighbors')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscancluster = DBSCAN(eps=0.3, min_samples=5)
dbscancluster.fit(x)
```

```python
clusters = dbscancluster.labels_

num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
print(f'Number of clusters: {num_clusters}')
```

```python
from sklearn.metrics import silhouette_score


if len(set(clusters)) > 1:
	score = silhouette_score(x, clusters)
	print(f'Silhouette Score: {score}')
else:
	print("Silhouette Score cannot be calculated. Number of clusters is less than 2.")
```
